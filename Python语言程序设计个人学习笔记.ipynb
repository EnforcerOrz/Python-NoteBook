### 1.2 实例1：温度转换
#### 问题:给出一摄氏度或华氏度，完成摄氏度和华氏度的相互转换
***
#### 问题分析
- 划分边界
    - 输入 带华氏度或摄氏度标志的温度值
    - 处理 根据温度标志选择适当的温度转换算法
    - 输出 带华氏度或摄氏度标志的温度值
- 设计算法  
根据定义，转换公式如下
$$ 
\begin{array}{l}{C=(F-32) / 1.8} \\ {F=C^{*} 1.8+32}\end{array}
 $$  
 ***
 #### 代码示例

```python
#TempConvert.py
Tempstr = input('请输入带有符号的温度值：') 
if Tempstr[-1] in ['F','f']:
    C=(eval(Tempstr[0:-1])-32)/1.8
    print('转换后的温度是{:.2f}C'.format(C))
elif Tempstr[-1] in ['C','c']:
    F=1.8*eval(Tempstr[0:-1])+32
    print('转换后的温度是{:.2f}F'.format(F))
else:
    print('输入格式错误')
```
